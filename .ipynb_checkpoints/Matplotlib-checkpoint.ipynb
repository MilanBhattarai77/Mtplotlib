```python
import matplotlib.pyplot as plt

students=[80,75,60,50,40]

l=["Python", "Java", "C", "HTML", "Scala"]

c=["r", "g", "y", "b", "c"]

e=[0.2, 0, 0, 0, 0]

plt.pie(students, colors=c, explode=e, labels=l, autopct="%0.2d%%", shadow=True, startangle=180, radius=1.1)

plt.title("Trending Languages")

plt.show()
```

```python
x=[1,2,3,4,5,6]

y=[90,86,70,69,78,88]

print(type(plt))

plt.xlabel("Parts")

plt.ylabel("Popularity")

plt.title("Harry Poter's  Parts and its popularity")

c=["red", "blue", "green", "orange", "pink", "yellow"]

plt.bar(x,y, color=c, edgecolor="black", alpha=0.5, label="Popularity")

plt.legend()

plt.show()
```

```python
y=[1,2,3,4,5,6]

x=[90,86,70,69,78,88]

c=["red", "blue", "green", "orange", "pink", "yellow"]

plt.xlabel("Popularity")

plt.ylabel("Seasons")
plt.barh(y,x, color=c, edgecolor="black", alpha=0.5, label="Popularity")

plt.legend()

plt.show()
```

```python
x=["Day1", "Day2", "Day3", "Day4"]

y=[30,40,60,20]

plt.barh(x,y)

plt.show()

for index,value in enumerate(y):
    plt.text(value, index, str(value))
```
