```python
data=[[30,50,60,45], [40,70,25,60]]

x=np.arange(4)

fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.bar(x, data[0], color="red", width=0.3, label="Women" )

ax.bar(x+0.3, data[1], color="blue", width=0.3, label="Man")

plt.legend()

plt.show()
```

```python
women=[20,50,80,40,70,55]

man=[40,70,30,60,70,35]

x=np.arange(6)

fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.bar(x, women, color="green", width=0.4, label="Women")

ax.bar(x, man, color="black", width=0.4, label="Man")

ax.set_xticks(x, ("D1", "D2", "D3", "D4", "D5", "D6"))

ax.set_yticks(np.arange(0, 120,20))

plt.legend()

plt.show()
```

```python
#fill between
```

```python
a=np.linspace(0, 2*3.14, 50)

b=np.sin(a)

plt.plot(a,b)

plt.fill_between(a, b, 0, where=(a>1.5)& (a<3.2), color="red")

plt.show()
```
